## **Primeiro gr√°fico Scatter**
## Expectativa de vida entre a Alemanha e Singapura


```python
# Definindo as vari√°veis (pa√≠s e ano)
pais1 = 'Germany'
pais2 = 'Singapore'

ano_inicio = 2010
ano_fim = 2020

# Selecionando os dados dos pa√≠ses + os anos que desejo visualizar
dados_pais1 = life_df[life_df['country'] == pais1].loc[:, str(ano_inicio):str(ano_fim)]
dados_pais2 = life_df[life_df['country'] == pais2].loc[:, str(ano_inicio):str(ano_fim)]
```

```python
# Criando o gr√°fico Scatter
plt.figure(figsize=(8, 6), facecolor='dimgray')

anos = range(ano_inicio, ano_fim + 1)

plt.scatter(anos, dados_pais1.values.flatten(), label=pais1, s= 100, color = 'red')   # Alemanha
plt.scatter(anos, dados_pais2.values.flatten(), label=pais2, s= 100, color = 'green') # Singapura

plt.title('Expectativa de vida da Alemanha e Singapura (2010-2020)')
plt.xlabel('Ano')
plt.ylabel('Expectativa de vida')
plt.legend()

plt.show()
```
